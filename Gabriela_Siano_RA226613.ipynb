# RMS Titanic

O RMS Titanic foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente "inafundável".

A embarcação partiu em sua viagem inaugural de Southampton para Nova Iorque em 10 de abril de 1912, no caminho passando em Cherbourg-Octeville na França e por Queenstown na Irlanda. Ele colidiu com um iceberg às 23h40min do dia 14 de abril e afundou na madrugada do dia seguinte matando mais de 1500 pessoas, sendo um dos maiores desastres marítimos em tempos de paz de toda a história. Seu naufrágio destacou vários pontos fracos de seu projeto, deficiências nos procedimentos de evacuação de emergência e falhas nas regulamentações marítimas da época. Comissões de inquérito foram instauradas nos Estados Unidos e no Reino Unido, levando a mudanças nas leis internacionais de navegação que permanecem em vigor mais de um século depois.

Os destroços do Titanic foram procurados por décadas até serem encontrados em 1985 por uma equipe liderada por Robert Ballard. Ele se encontra a 3843 m de profundidade e a 650 km ao sudeste de Terra Nova no Canadá. Sua história e naufrágio permaneceram no imaginário popular durante décadas, levando a produção de vários livros e filmes a seu respeito, mais notavelmente o filme Titanic de 1997. Até hoje o Titanic permanece como um dos navios mais famosos da história, com seus destroços atraindo várias expedições de exploração ao longo dos anos.

4 - Cálculos envolvendo colunas numéricas com dados faltantes podem sofrer impacto. É possível afirmar se há dados faltantes no dataset? Caso positivo, quais e quantos seriam esses dados? Preencha os dados faltantes de forma que não influenciem em operações futuras.

```python
#Avaliando quantos dados estão ausentes em cada coluna
print("Dados Faltantes:")
tit.isnull().sum().sort_values(ascending=False) # Temos dados faltantes em 'Age', 'Cabin' e ' Embarked
```

```python
from scipy.stats import mode
#Existem 3 opções para o caso de ausencia de dados
#1- Excluir linhas que possuem dados ausentes 
#2- Preencher valores ausentes com a mediana da coluna
#3- Usar o valor mais frequente - para casos não numericos

# identificar o valor mais frequente
#Avaliar dados preenchidos no cabin
print(f"O Valor mais frequente é {mode(tit['Cabin']).mode[0]}\n")


# identificar Mediana de idade
mediana_Age = tit.Age.median()
print(f"A mediana da idade é {mediana_Age}\n")

#Verificar o total de dados
print(f'O dataset possui {linhas} linhas e {colunas} colunas.\n')
```

```python
# preencher valores ausentes com maior frequencia / mediana
frequencia_Cabin = mode(tit['Cabin']).mode[0]
tit.Cabin.fillna(frequencia_Cabin, inplace=True)
tit.fillna(mediana_Age, inplace=True)

# Como se trata de apenas duas linhas (0,22% de informação), foi optado por exclusão dos vazios em Embarked
tit.dropna(subset=['Embarked'], inplace=True)


# verificando se ainda há valor ausente
print("Verificando se ainda há valores ausentes em cada coluna")
tit.isnull().sum().sort_values(ascending=False)
```

## Adequando o DataSet

5 - Uma vez que algumas colunas não serão utilizadas, eventualmente é melhor excluí-las para que não interfiram na análise. Assim, exclua do dataset as colunas Sibsp, Parch e Ticket.

6 - Renomear as colunas restantes para a lingua portuguesa, utilizando os seguintes nomes de colunas: IdPassageiro, Sobreviveu, Classe, Nome, Sexo, Idade, Tarifa, Cabine e Embarque.

12 - Proporcionalmente, sobreviveram mais homens ou mais mulheres? Cite as proporções.

```python
#Verificar proporções de homens e mulheres que sobreviveram em relação ao total de sobreviventes

total_sobreviveram = tit[tit['Sobreviveu'] == 'Sim']

mulheres_sobreviveram = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Mulher')]

homem_sobreviveram = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Homem')]

porcentagem_mulheres= (mulheres_sobreviveram.shape[0]/total_sobreviveram.shape[0])*100
porcentagem_homens= (homem_sobreviveram.shape[0]/total_sobreviveram.shape[0])*100

if porcentagem_mulheres > porcentagem_homens:
    print("Proporcionalmente, sobreviveram mais mulheres. Seguem as proporções de homens e mulheres que sobreviveram: ")

else:
    print("Proporcionalmente, sobreviveram mais homens. Seguem as proporções de homens e mulheres que sobreviveram: ")

print(f'Homens: {round(porcentagem_homens, 1)}%\nMulheres: {round(porcentagem_mulheres,1)}%')
```

13 - Levando-se em consideração a idade dos passageiros, qual a idade e quantidade de pessoas com o maior número de mortos?

```python
#Criando um dataset apenas com as linhas de não sobreviventes:
mortos = tit[tit.Sobreviveu=='Não']

#Identificando o valor de idade mais frequente no novo dataset:
moda = int(mortos.Idade.mode())

#Criando um dataset apenas com os valores de idade que correspondem à moda:
mortos_moda = mortos[mortos['Idade']==moda]

#Calculando a quantidade de linhas no novo dataset:
quantidade = mortos_moda.shape[0]

print(f'A idade com o maior número de mortos foi {moda} anos, com {quantidade} mortos.')
```

15 - Levando-se em consideração passageiros prioritários (mulheres e crianças de até 15 anos independente do sexo) qual a proporção de sobreviventes por sexo?

```python
#Filtrando a quantidade de Mulheres e Crianças
quant_mulher_crianca = tit [
                (tit['Sexo'] == 'Mulher') | 
                (tit['Idade'] <= 15)                     
        ]
#Filtrando os Sobreviventes        
sobrev_mulher_crianca = quant_mulher_crianca[
                          quant_mulher_crianca['Sobreviveu'] == 'Sim'
]
#Filtrando só as mulheres
mulher_sobreviventes = sobrev_mulher_crianca [
    sobrev_mulher_crianca['Sexo'] == 'Mulher'
]  
#Filtrando as crianças do sexo masculino
crianca_sobreviventes = sobrev_mulher_crianca [
    sobrev_mulher_crianca['Sexo'] == 'Homem'
]

#Percentual de mulheres
perc_total_mulheres = mulher_sobreviventes.shape[0] * 100 / sobrev_mulher_crianca.shape[0]
 
#Percentual de Homens
perc_total_homens = crianca_sobreviventes.shape[0] * 100 / sobrev_mulher_crianca.shape[0]

print(f'Mulheres sobreviventes: {round(perc_total_mulheres, 2)}% \nHomens sobreviventes (crianças): {round(perc_total_homens,2)}%')      
```

20 - Calcule a probabilidade condicional de uma pessoa sobreviver, dado seu sexo e a classe em que estava viajando:
- P(S= true | G=female,C=1)
- P(S= true | G=female,C=2)
- P(S= true | G=female,C=3)
- P(S= true | G=male,C=1)
- P(S= true | G=male,C=2)
- P(S= true | G=male,C=3)

```python
#Calculando o total de sobreviventes:
sobreviventes_total = sobreviventes_total = tit[tit['Sobreviveu'] == 'Sim'].shape[0]

#Calculando mulheres e homens sobreviventes por classe:
classe_1_mulher_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') &(tit['Sexo'] == 'Mulher')&(tit['Classe'] == 1)]
classe_2_mulher_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Mulher')&(tit['Classe'] == 2)]
classe_3_mulher_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Mulher')&(tit['Classe'] == 3)]

classe_1_homem_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') &(tit['Sexo'] == 'Homem')&(tit['Classe'] == 1)]
classe_2_homem_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Homem')&(tit['Classe'] == 2)]
classe_3_homem_sobreviventes = tit[(tit['Sobreviveu'] == 'Sim') & (tit['Sexo'] == 'Homem')&(tit['Classe'] == 3)]

# Calculando a probabilidade de mulheres sobreviverem por classe:
probabilidade_mulher_classe_1 = ((classe_1_mulher_sobreviventes.shape[0] / sobreviventes_total) * 100)
probabilidade_mulher_classe_2 = ((classe_2_mulher_sobreviventes.shape[0] / sobreviventes_total) * 100)
probabilidade_mulher_classe_3 = ((classe_3_mulher_sobreviventes.shape[0] / sobreviventes_total) * 100)

# Calculando a probabilidade de homens sobreviverem por classe:
probabilidade_homem_classe_1 = ((classe_1_homem_sobreviventes.shape[0] / sobreviventes_total) * 100)
probabilidade_homem_classe_2 = ((classe_2_homem_sobreviventes.shape[0] / sobreviventes_total) * 100)
probabilidade_homem_classe_3 = ((classe_3_homem_sobreviventes.shape[0] / sobreviventes_total) * 100)

# Mostrando a probabilidade em Data Frame
probabilidade = pd.DataFrame({'Sexo': ['Mulher', 'Homem'],
                              '1ª Classe (%)': [round(probabilidade_mulher_classe_1, 2), round(probabilidade_homem_classe_1, 2)],
                              '2ª Classe (%)': [round(probabilidade_mulher_classe_2, 2), round(probabilidade_homem_classe_2, 2)],
                              '3ª Classe (%)': [round(probabilidade_mulher_classe_3, 2), round(probabilidade_homem_classe_3, 2)]                            })
probabilidade.head()
```

```python
probabilidade.plot(kind='bar', subplots=True, layout=(3, 3), figsize=(10, 10), sharey=True)
```
